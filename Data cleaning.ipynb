Importing library required for task

```python
import pandas as pd
import re
from datetime import datetime
```

```python
import warnings
warnings.filterwarnings("ignore")
```

reading data into whatsapp df

```python
whatsapp_df = pd.read_fwf('WhatsApp Chat with M tech Nirma.txt', header = None)
whatsapp_df
```

```python
whatsapp_df = whatsapp_df[[0,1]]
whatsapp_df.columns=["Date","Message"]
whatsapp_df
```

```python
whatsapp_df.dropna(how="any",inplace=True)
whatsapp_df.reset_index(drop=True,inplace=True)
```

```python
# https://stackoverflow.com/questions/4709652/python-regex-to-match-dates
# https://www.tutorialspoint.com/Why-do-we-use-re-compile-method-in-Python-regular-expression
dup_message = [] ##index whose message we have appended to previous message. 
for i in range(len(whatsapp_df)):
    date = re.compile(r'\d+/\d+/\d+')
    if date.findall(whatsapp_df.iloc[i,0]):
        continue
    else:
        temp = whatsapp_df.loc[i,"Date"] + whatsapp_df.loc[i,"Message"]
        whatsapp_df.loc[i-1,"Message"] = whatsapp_df.loc[i-1,"Message"] + str(" ")+ temp
        dup_message.append(i)
```

```python
# https://stackoverflow.com/questions/14661701/how-to-drop-a-list-of-rows-from-pandas-dataframe
whatsapp_df_no_dup_message = whatsapp_df.drop(whatsapp_df.index[dup_message])
whatsapp_df_no_dup_message.reset_index(drop=True,inplace=True)
```

```python
whatsapp_df_no_dup_message
```

```python
for i in range(len(whatsapp_df_no_dup_message["Message"])):
    message = str(whatsapp_df_no_dup_message["Message"].iloc[i]).split("-",2)
    whatsapp_df_no_dup_message.loc[i,"Time"] = message[0]
    whatsapp_df_no_dup_message.loc[i,"Message"] = message[1]
```

```python
whatsapp_df_no_dup_message
```

Handling Time

```python
for i in range(len(whatsapp_df_no_dup_message["Time"])):
    time  = []
    time.append(whatsapp_df_no_dup_message.loc[i,"Time"].split(":")[0])
    time.extend(whatsapp_df_no_dup_message.loc[i,"Time"].split(":")[1].split(" "))
    if time[2] == "pm" and int(time[0])!=12:
        time[0] = str(int(time[0])+12)
        whatsapp_df_no_dup_message.loc[i,"time"]=":".join(time[:2])
    else:
        whatsapp_df_no_dup_message.loc[i,"time"]=":".join(time[:2])
whatsapp_df_no_dup_message.drop("Time",axis=1,inplace=True)
whatsapp_df_no_dup_message
```

adding sender in sender columns

```python
for i in range(len(whatsapp_df_no_dup_message)):
    message = whatsapp_df_no_dup_message.loc[i,"Message"]
    string = message.split(":")
    if len(string)>1:
        whatsapp_df_no_dup_message.loc[i,"Sender"] = string[0]
        whatsapp_df_no_dup_message.loc[i,"Message"] = " ".join(string[1:])
print(whatsapp_df_no_dup_message)    
```

Removing all rows where sender is " "(messages send by whatsapp when someone leave or join group)

```python
whatsapp_df_no_dup_message.dropna(how="any",inplace=True)
whatsapp_df_no_dup_message.reset_index(drop=True,inplace=True)
```

removing commas in date

```python
whatsapp_df_no_dup_message.loc[:,"Date"] = whatsapp_df_no_dup_message.loc[:,"Date"].replace(',','', regex=True)
whatsapp_df_no_dup_message
```

Converting datetime string object into Datetime

```python
#https://stackoverflow.com/questions/33098383/merge-multiple-column-values-into-one-column-in-python-pandas
whatsapp_df_no_dup_message = whatsapp_df_no_dup_message.assign(datetime = whatsapp_df_no_dup_message.Date.astype(str)+" "+whatsapp_df_no_dup_message.time.astype(str))
whatsapp_df_no_dup_message.drop(["Date","time"],axis=1,inplace=True)
```

```python
def date_convert(date_str):
    return datetime.strptime(date_str,"%d/%m/%y %H:%M")
whatsapp_df_no_dup_message["datetime"] = whatsapp_df_no_dup_message.apply(lambda row: date_convert(row["datetime"]),axis = 1)
print(type(whatsapp_df_no_dup_message.loc[1,"datetime"]))
```
